```python
from PIL import Image
import requests
import matplotlib.pyplot as plt
import cv2
from datetime import datetime
import numpy as np
import random
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
url = "https://i.postimg.cc/cHdcX3jV/uno.jpg"
r=requests.get(url)
open('image.jpg','wb').write(r.content)
image=Image.open('image.jpg')
fig = plt.figure(figsize=(10, 7))#configurar ptl
plt.imshow(image)
plt.axis('off')
plt.title("Original")
```

```python
img = cv2.imread('image.jpg')[:,:,::-1]
img2 = img

height, width, channels = img.shape
# Number of pieces Horizontally
W_SIZE  = 3
# Number of pieces Vertically to each Horizontal
H_SIZE = 3
img_cuts=[]
for ih in range(H_SIZE ):
   for iw in range(W_SIZE ):

      x = width/W_SIZE * iw
      y = height/H_SIZE * ih
      h = (height / H_SIZE)
      w = (width / W_SIZE )
      img = img[int(y):int(y+h), int(x):int(x+w)]
      img_cuts.append(img)
      img = img2
random.shuffle(img_cuts)
fig = plt.figure(figsize=(10, 7))#configurar ptl
i=0
for ih in range(H_SIZE):
  for iw in range(W_SIZE):
    fig.add_subplot(H_SIZE, W_SIZE, i+1)#configurar subplot
    plt.imshow(img_cuts[i])
    plt.axis('off')
    plt.title(str(i))
    i += 1
```
